```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook() #we should write this line else it will not run the graphs inside your notebook for bokeh
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p = figure(title="test flower")
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
output_file('test.html')
p = figure(title="test flower")
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,2,1]
output_file('test.html')
p = figure(title='Line Graph')
p.line(x,y)
show(p)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,2,1]
output_file('test.html')
p = figure(title='Line Graph')
p.scatter(x,y,size=5)
show(p)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,2,1]
output_file('test.html')
p = figure(title='Line Graph')
p.circle(x,y,size=10,fill_color='red')
show(p)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,2,1]
output_file('test.html')
p = figure(title='Line Graph')
p.circle(x,y,size=10,fill_color='red',legend_label='Data Point')
show(p)
```
